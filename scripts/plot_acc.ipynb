```python
import pandas as pd
import seaborn as sns
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
s5 = pd.read_csv('s100_n5_c0.tsv',sep='\t',comment='#')['avg. acc']
s10 = pd.read_csv('s100_n10_c0.tsv',sep='\t',comment='#')['avg. acc']
s15 = pd.read_csv('s100_n15_c0.tsv',sep='\t',comment='#')['avg. acc']
s20 = pd.read_csv('s100_n20_c0.tsv',sep='\t',comment='#')['avg. acc']
s25 = pd.read_csv('s100_n25_c0.tsv',sep='\t',comment='#')['avg. acc']
s30 = pd.read_csv('s100_n30_c0.tsv',sep='\t',comment='#')['avg. acc']
df = pd.concat([100*_ for _ in [s5,s10,s15, s20,s25,s30]], axis=1)
df.columns = ['T=5', 'T=10', 'T=15', 'T=20', 'T=25', 'T=30']
```

```python
# figure size in inches
rcParams['figure.figsize'] = 8,6

ax = sns.swarmplot(data=df)
ax = sns.boxplot(data=df, boxprops={'facecolor':'None'}, whiskerprops={'linewidth':0}, showcaps=False)

_=ax.set_xticklabels(df.columns,size=25,rotation=45)
_=ax.set_yticklabels([f"{_:.0f}" for _ in ax.get_yticks()],size=24)
_=ax.set_title('Number of inputs vs avg. retrieval accuracy', size=24)

plt.tight_layout()
plt.savefig('swarmplot.png')
```

### Plot accuracy as a function of sparsity

```python
rcParams['figure.figsize'] = 10,6

p10 = pd.read_csv('s100_n15_p10.tsv',sep='\t',comment='#')['avg. acc']
p20 = pd.read_csv('s100_n15_p20.tsv',sep='\t',comment='#')['avg. acc']
p30 = pd.read_csv('s100_n15_p30.tsv',sep='\t',comment='#')['avg. acc']
p40 = pd.read_csv('s100_n15_p40.tsv',sep='\t',comment='#')['avg. acc']
p50 = pd.read_csv('s100_n15_p50.tsv',sep='\t',comment='#')['avg. acc']
p60 = pd.read_csv('s100_n15_p60.tsv',sep='\t',comment='#')['avg. acc']
p70 = pd.read_csv('s100_n15_p70.tsv',sep='\t',comment='#')['avg. acc']
p80 = pd.read_csv('s100_n15_p80.tsv',sep='\t',comment='#')['avg. acc']
p90 = pd.read_csv('s100_n15_p90.tsv',sep='\t',comment='#')['avg. acc']

df = pd.concat([100*_ for _ in [p10,p20,p30,p40,p50,p60,p70,p80,p90]], axis=1)
df.columns = ['P=10','P=20','P=30','P=40','P=50','P=60','P=70','P=80','P=90']

ax = sns.boxplot(data=df, boxprops={'facecolor':'None'}, whiskerprops={'linewidth':0}, showcaps=False)
ax = sns.stripplot(data=df)
_=plt.title('Sparsity (P) vs avg. retrieval accuracy', size=24)

_=ax.set_xticklabels(df.columns,size=25,rotation=45)
_=ax.set_yticklabels([f"{_:.0f}" for _ in ax.get_yticks()],size=24)
_=ax.set_title('Sparsity of 15 input vectors vs retrieval accuracy', size=24)

plt.tight_layout()
plt.savefig('stripplot.png')
```
